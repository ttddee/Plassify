```python
import os

import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
train_folder = 'SLoPP and SLoPP-E\SLoPP'
```

```python
# Load data from a single text file
def load_file(path):
    # Load and crop the array, because this data set is not uniform
    data = np.loadtxt(path)[0:743,:]
    
    # Standard Normal Variate?
    
    # Normalize Absorbance -- can go into plotting function?
    norm_data = np.copy(data)
    norm_data[:,1] = norm_data[:,1] / norm_data[:,1].max(axis=0)
    
    #print(np.shape(norm_data))
    
    return norm_data
```

```python
# Load all files in folder and create X array
num_samples = 0
data_list = []

for path, subdirs, files in os.walk(train_folder):
    for name in files:
        data_list.append(load_file(os.path.join(path, name))[:,1])
        num_samples += 1

X = np.stack(data_list)

print("{} samples loaded.".format(num_samples))
```

```python
# Load labels and create y array
y = []

for path, subdirs, files in os.walk(train_folder):
    for name in files:
        y.append(os.path.split(path)[-1])

print("{} labels loaded.".format(len(y)))
```

```python
# Separate train and test samples
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2)
```

```python
# Linear Discriminant Analysis
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X, y)
Xlda = lda.fit_transform(X,y)
```

```python
# Load data
filename = "SLoPP and SLoPP-E\SLoPP\Train\Acrylic\Acrylic 1. Green Yarn.txt"

input_data = load_file(filename)

print(input_data)
```
